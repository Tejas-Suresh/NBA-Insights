```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
dat=pd.read_csv('NBA_Data.csv')
dat=dat.dropna()
dat.head()
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
mean=np.mean(dat["NBA_ppg"])
meang=np.mean(dat["NBA_g_played"])
```

```python
dat.plot.scatter(x="NBA_g_played",y="NBA_ppg");
plt.scatter(x=meang,y=mean,color="r",alpha=0.5,s=100,label="Mean")
plt.legend()
```

```python
plt.hist(dat["NBA_ppg"],bins=20)

plt.axvline(mean,color="red",label="Mean PPG")
plt.xlabel("NBA_ppg")
plt.ylabel("Number of players")
plt.legend()
```

```python
ax=dat.plot.scatter("NCAA_ppg","NBA_ppg" , c="NBA_g_played", colormap='viridis',figsize=(10,10),title="Relation between NBA ppg and NCAA ppg")
ax.set_xlabel("NCAA")
plt.draw()
```

```python
l=list()
for i in dat["NBA_g_played"]:
    if i<200:
        l.append(0)
    elif i>200 and i<600:
        l.append(1)
    elif i>600 and i<1000:
        l.append(2)
    elif i>1000 and i<1400:
        l.append(3)
    elif i>1400:
        l.append(4)
dat["gclass"]=l
replace_dict={0:"Low",1:"Medium",2:"Good",3:"Very good",4:"High"}
dat["gclass"].replace(to_replace=replace_dict, inplace=True)
dat.head()
```

```python
dlow=dat[dat["gclass"]=="Low"]
dmed=dat[dat["gclass"]=="Medium"]
dgood=dat[dat["gclass"]=="Good"]
dvg=dat[dat["gclass"]=="Very Good"]
dhigh=dat[dat["gclass"]=="High"]
dlow


```

```python
df_list = [dlow,dmed,dgood,dhigh]
l=["Low(less than 200 career games)","Medium(between 200-600 career games)","Good(between 600-1400 career games)","High(greater than 1400 career games)"]
fig, axes = plt.subplots(2,2)

# plot counter
count=0
for r in range(2):    
    for c in range(2):
        df_list[count].plot.scatter("WT","NBA_ppg",c="NCAA_games",colormap='viridis',figsize=(15,15),ax=axes[r,c],title=l[count])
        count=count+1
fig.suptitle('A relation between weight, ppg and NCAA games played for NBA players(classified by career games)',fontsize=15)    
plt.tight_layout(pad=5)
```
