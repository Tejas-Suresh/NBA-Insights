```python
threeptmed = pd.Series.median(compd['NBA__3ptpct'].dropna())
threeptmed
```

```python
threeptmean = pd.Series.mean(compd['NBA__3ptpct'].dropna())
threeptmean
```

```python
threeptstd = pd.Series.std(compd['NBA__3ptpct'].dropna())
threeptstd
avggplay=pd.Series.mean(compd["NBA_g_played"].dropna())
```

```python
cond1=compd['NBA__3ptpct']>(threeptmean)
cond2=compd["position"]=='G'
cond3=compd["NBA_g_played"]>avggplay
cond4=compd["NBA__3ptapg"]>pd.Series.mean(compd["NBA__3ptapg"].dropna())
compd.loc[cond1&cond2&cond3&cond4,:]
```
